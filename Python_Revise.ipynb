#### Importing Pandas
It is a python library used to organise data in tabular format and can be used to manipulate and anlyse data. It can be used to replace excel iin perfroming ETL ( Extract, Transforming and loading of data)


```python
import pandas as pd  ## It is case sensitive. Don't use capital letters in your code
```

```python
fruits = pd.Series(['Apples', 'Avocados', 'Bananas','Blueberries', 'Broccoli' ,'Capsicums', 'Carrots', ' Cucumbers', 'Grapes', 'Lemon',
          'Lettuce', 'Limes', 'Mandarins', 'Mangoes', 'Mushrooms', 'Onions', 'Oranges', 'Pineapples', 'Potatoes', 'Rockmelons',
          'Strawberries', 'Tomatoes', 'Watermelons'], name='Fruits')
avgprices= pd.Series([2.3, 3.5, 5.0, 11.3, 2.2, 2.5, 3.2, 4.7, 14.5, 3.5, 3.4, 2.2, 2.6, 9.8, 3.3, 5.8, 8.5, 6.4, 10.5, 6.8, 5.6, 7.8, 9.3], name='avg_prices')
```

```python
avgprices
```

```python
df = pd.DataFrame({'Fruits' : fruits, 'Avg_Prices' : avgprices}) # combining both series to form a dataframe
print(df)
```

### .loc attribute

#### Selection with .loc follows this structure:
#### df.loc[row_label(s), column_label(s)]

```python
df2.loc[0, "Commodity"]
```

### .iloc attribute
It is used for position based indexing. It follows the same structure as .loc

#### df.iloc[row_position_index, column_position_index]

```python
df2.iloc[2,4]
```

### Creating Custom Index
We use the parameter 'index' when creating the dataframe

```python
df_custom = pd.DataFrame({'Name': ['Alice', 'Bob', 'Claire'],
                        'Age' : [25, 30, 22]
                         })                        
                       # index=['ID001', 'ID002', 'ID003'])
#df_custom.index
```

```python
df_custom = pd.DataFrame({'Name': ['Alice', 'Bob', 'Claire'],
                          'Age': [25, 30, 22]}) ,
                         index=['ID001', 'ID002', 'ID003'])

#df_custom.index
```
